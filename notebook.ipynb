```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
dates = ["lpep_pickup_datetime","lpep_dropoff_datetime"]
```

```python
df = pd.read_parquet("https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet",engine="pyarrow")
for date in dates:
    df[date] = pd.to_datetime(df[date]).dt.date
df.head()
```

```python
df.shape[0]
```

```python
engine = create_engine("postgresql://root:root@localhost:5432/ny_taxi")
df.head(0).to_sql(name="green_trips",con=engine,if_exists="replace",index=False)
df.to_sql(name="green_trips",con=engine,if_exists="append",index=False)
```
